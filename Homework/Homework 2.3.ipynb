### Homework 2.3 - 2D Temperature Distribution

Solve 2D heat equation for the following initial conditions:

$$\begin{array}{l}
u(x,0) = \sin (2\pi x)\\
u(x,1) = \sin (2\pi x)\\
u(0,y) = \sin (3\pi y)\\
u(1,y) = \sin (\pi y)
\end{array}$$

where $u(x, y)$ represent the steady state temperature distribution on 2D plane.
 
Plot the 2D heat map of the solution.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
etol = 1e-5
maxiter = 1000
dim = 100
x = np.linspace(0, 1, dim)
y = np.linspace(0, 1, dim)

U = np.zeros((dim,dim))


U[:, 0] = np.sin(2*np.pi*x)
U[:, dim-1] = np.sin(2*np.pi*x)
U[0, :] = np.sin(3*np.pi*y)
U[dim-1, :] = np.sin(np.pi*y)

done = False
count = 0
while done is False:
    Uold = U.copy()
    for i in range(1, dim-1):
        for j in range(1, dim-1):
            U[i][j] = 0.25*(U[i+1][j] + U[i-1][j] + U[i][j+1] + U[i][j-1])
            
    error = np.sum(np.abs(Uold-U))
    count += 1
    done = ((error/dim**2 < etol) or (count > maxiter))
    
plt.imshow(U, cmap="hot")
plt.show()
```
