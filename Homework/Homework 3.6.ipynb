The y that we get from wavfile is Time-Domain data and we want to transform time-domain wave to frequency-domain wave by fourier transform.

Then we use rfft function to get Frequency-Domain view of original wave.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq, irfft
```

```python
from scipy.io import wavfile
freq, y = wavfile.read('images/record2.wav')
dt = 1/freq
t = np.linspace(0, (len(y)-1)*dt, len(y))
plt.plot(t, y, label='Original')
plt.title('Time Domain')
plt.legend()
```

```python
n = len(t)
yf = rfft(y)
xf = rfftfreq(n, dt)

plt.plot(xf, np.abs(yf))
```

#### 1
roll the array to increase the pitch.

```python
new_yf1 = np.roll(yf, 500)

plt.plot(xf, np.abs(new_yf1))
```

```python
new_f1 = irfft(new_yf1)
```

```python
wavfile.write('change_voice1.wav', freq, np.array(new_f1, dtype=np.int16))
```
